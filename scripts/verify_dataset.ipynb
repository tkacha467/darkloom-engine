```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("f:\\darkloom\\darkloom-engine\\scripts\\metadata\\event_001_pairs.csv")

print("Total rows:", len(df))
df.head()
```

```python
import rawpy
import matplotlib.pyplot as plt
from PIL import Image
import os

base_path = "H:/dataset"

for i in range(5):

    row = df.iloc[i]

    raw_path = os.path.join(base_path, row['raw_path'])
    edited_path = os.path.join(base_path, row['edited_path'])

    print(raw_path)
    print(edited_path)

    # Read RAW image
    with rawpy.imread(raw_path) as raw:
        raw_img = raw.postprocess()

    # Read edited JPG
    edited_img = Image.open(edited_path)

    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(raw_img)
    plt.title("RAW")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(edited_img)
    plt.title("EDITED")
    plt.axis("off")

    plt.show()
```
