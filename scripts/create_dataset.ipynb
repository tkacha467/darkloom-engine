```python
import os
import pandas as pd
```

```python
base_path = "H:\\dataset\\event_001"

raw_folder = os.path.join(base_path, "raw")
edited_folder = os.path.join(base_path, "edited")
xmp_folder = os.path.join(base_path, "xmp")
selected_folder = os.path.join(base_path, "selected")
```

```python
data = []

raw_files = os.listdir(r"H:\dataset\event_001\raw")
edited_files = os.listdir(r"H:\dataset\event_001\edited")
xmp_files = os.listdir(r"H:\dataset\event_001\xmp")

# convert to lowercase sets for matching
edited_set = {f.lower(): f for f in edited_files}
xmp_set = {f.lower(): f for f in xmp_files}

for file in raw_files:
    if file.lower().endswith(".arw"):
        
        name = os.path.splitext(file)[0]
        
        edited_name = name + ".jpg"
        xmp_name = name + ".xmp"
        
        if edited_name.lower() in edited_set:
            
            raw_path = f"event_001/raw/{file}"
            edited_path = f"event_001/edited/{edited_set[edited_name.lower()]}"
            
            xmp_path = ""
            if xmp_name.lower() in xmp_set:
                xmp_path = f"event_001/xmp/{xmp_set[xmp_name.lower()]}"
            
            data.append([raw_path, edited_path, xmp_path])

print("Total pairs:", len(data))
```

```python
df = pd.DataFrame(data, columns=[
    "raw_path",
    "edited_path",
    "xmp_path"
])

df.head()
```

```python
# create metadata folder if not exists
os.makedirs("metadata", exist_ok=True)

# save CSV
df.to_csv("metadata/event_001_pairs.csv", index=False)

# save Excel
df.to_excel("metadata/event_001_pairs.xlsx", index=False)

print("✅ Files saved in metadata folder")
```
